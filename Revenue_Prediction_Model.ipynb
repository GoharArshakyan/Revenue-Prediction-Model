```python
#import necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#upload the data
rev_data = pd.read_csv("C://Users/user/Desktop/Internship_Task/data_scientist_intern_revenue_model.csv", na_values = ' -   ')
print(rev_data.head(5))
```

```python
#check the dimensions of the data
rev_data.shape
```

```python
#check for data types
rev_data.info()
```

```python
#change object types to floats
rev_data["Revenue_2015 "] = rev_data["Revenue_2015 "].str.replace(",", "").astype(float)
rev_data["Revenue_2016"] = rev_data["Revenue_2016"].str.replace(",", "").astype(float)
rev_data["Revenue_2017"] = rev_data["Revenue_2017"].str.replace(",", "").astype(float)
rev_data["Revenue_2018"] = rev_data["Revenue_2018"].str.replace(",", "").astype(float)
rev_data["Revenue_2019"] = rev_data["Revenue_2019"].str.replace(",", "").astype(float)
```

```python
#check summary statistics
rev_data.describe()
```

```python
#how many values are missing in each column
rev_data.isnull().sum()
```

```python
#percentage of missing values in each column
rev_data.isnull().sum() * 100/len(rev_data)
```

```python
#filling in missing values with the mean of the column
rev_data["Revenue_2015 "] = rev_data["Revenue_2015 "].fillna(rev_data["Revenue_2015 "].mean())
rev_data["Revenue_2016"] = rev_data["Revenue_2016"].fillna(rev_data["Revenue_2016"].mean())
rev_data["Revenue_2017"] = rev_data["Revenue_2017"].fillna(rev_data["Revenue_2017"].mean())
rev_data["Revenue_2018"] = rev_data["Revenue_2018"].fillna(rev_data["Revenue_2018"].mean())
rev_data["Revenue_2019"] = rev_data["Revenue_2019"].fillna(rev_data["Revenue_2019"].mean())
rev_data["Employees_2017"] = rev_data["Employees_2017"].fillna(rev_data["Employees_2017"].mean())
rev_data["Employees_2018"] = rev_data["Employees_2018"].fillna(rev_data["Employees_2018"].mean())
rev_data["Employees_2019"] = rev_data["Employees_2019"].fillna(rev_data["Employees_2019"].mean())
rev_data["Employees_2020"] = rev_data["Employees_2020"].fillna(rev_data["Employees_2020"].mean())
rev_data["Revenue_2020(To predict)"] = rev_data["Revenue_2020(To predict)"].fillna(0)
```

```python
#verifying that there is no missing values
rev_data.isnull().sum()
```

```python
#checking for duplicates
duplicate = rev_data.duplicated()
print(duplicate.sum())
```
